```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
merged=movies.merge(credits,left_on='id', right_on ='movie_id')
```

```python
merged.head()
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
df=merged[['movie_id','original_title','genres','overview','keywords','cast','crew']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.shape
```

```python
df['keywords'] = df['keywords'].apply(convert)
df.head()
```

```python

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
        else:
            break
    return L
```

```python
df['cast'] = df['cast'].apply(convert3)
df.head()
```

```python
def fetchDir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew'] = df['crew'].apply(fetchDir)
df.head()
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
df.head()
```

```python
df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df.head()
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head()
```

```python
new_df=df[['movie_id','original_title','tags']]
```

```python
new_df.shape
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(txt):
    y=[]
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
from IPython import get_ipython
if get_ipython() is not None:
    get_ipython().ast_node_interactivity = 'all'
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
features=cv.get_feature_names_out()
```

```python
list(features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    dist=similarity[movie_index]
    recommended_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]

    for i in recommended_list:
        
        print(new_df.iloc[i[0]].original_title)
```

```python
recommend('Batman Begins')
```
